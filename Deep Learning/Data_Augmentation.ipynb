```python
from keras.preprocessing.image import ImageDataGenerator,array_to_img, load_img,img_to_array
```

```python
datagen=ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.20,
    height_shift_range=0.20,
    shear_range=0.20,
    zoom_range=0.20,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
# load image
img_path='/content/drive/MyDrive/Tutort/datasets/Predict_Image/Image_01.jpg'
save_image_path='/content/drive/MyDrive/Tutort/datasets/Augmented_Image'
img=load_img(img_path,target_size=(180,180))
```

```python
img
```

```python
# Convert Image to numpy array
x=img_to_array(img)
print(x.shape)
```

```python
# change to 4 dim
x=x.reshape((1,)+x.shape)
print(x.shape)
```

```python
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir=save_image_path,save_prefix='dog',save_format="jpeg"):
  i+=1
  if i==20:
    break
```
