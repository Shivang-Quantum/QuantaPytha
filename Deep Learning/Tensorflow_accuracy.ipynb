```python
# Defining the Data
df=pd.read_csv('/content/sample_data/diabetes.csv')
```

```python
# Splitting the data
x_train=df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].values
y_train=df.Outcome.values

#df.Outcome.values
```

```python
x_train
```

```python
# Build Architecture

model=tf.keras.Sequential()

#hidden Layer 1
#model.add(tf.keras.layers.Dense(10))
model.add(tf.keras.layers.Dense(10,activation='relu',kernel_initializer='he_uniform'))

#hidden Layer 2
#model.add(tf.keras.layers.Dense(10))
model.add(tf.keras.layers.Dense(10,activation='sigmoid'))

#Output Layer [The layer that is at the end]
model.add(tf.keras.layers.Dense(1))
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error',metrics=['Accuracy'])
```

```python
model.fit(x_train,y_train,epochs=20)
```
